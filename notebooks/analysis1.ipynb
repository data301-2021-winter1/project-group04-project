## Clean Data

```python
mcu1= mcu.copy().drop(columns=['full_name', 'occupation', 'type_race', 'has_jump', 'has_illusions', 'alter_egos', 'first_appearance', 'relatives', 'eye_color', 'hair_color', 'skin_color', 'img', 'has_electrokinesis', 'has_energy_constructs', 'has_matter_manipulation', 'has_telepathy_resistance', 'has_enhanced_hearing', 'has_dimensional_travel', 'has_element_control', 'has_size_changing', 'has_fire_resistance', 'has_fire_control', 'has_dexterity', 'has_reality_warping', 'has_energy_beams', 'has_shapeshifting', 'has_heat_resistance', 'has_heat_resistance', 'has_energy_absorption', 'has_cold_resistance', 'has_magic', 'has_toxin_and_disease_resistance', 'has_regeneration', 'has_immortality', 'has_teleportation', 'has_force_fields', 'has_energy_manipulation', 'has_longevity', 'has_weapon-based_powers', 'has_energy_blasts', 'has_enhanced_senses', 'has_invulnerability', 'has_stealth', 'has_marksmanship', 'has_flight', 'has_accelerated_healing', 'has_weapons_master', 'has_telepathy', 'has_telekinesis', 'has_self-sustenance'])
```

```python
mcu1 = mcu1[mcu1['intelligence_score'] > 0]
```

```python
mcu1 = mcu1[mcu1['strength_score'] > 0]
```

## Process Data

```python
sum_column = mcu1["intelligence_score"] + mcu1["strength_score"]
mcu1["intelligence_strength_sum"] = sum_column
```

```python
ax= mcu1["strength_score"].plot(kind='hist', bins=20, figsize=(15,6), facecolor='red',edgecolor='black')
ax.set_title("Frequency of Strength Scores")
```

```python
ax= mcu1["intelligence_score"].plot(kind='hist', bins=20, figsize=(15,6), facecolor='blue',edgecolor='black')
ax.set_title("Frequency of Intelligence Scores")
```

### Histogram Analysis

**These histograms show the frequency in which heroes recieved certain intelligence scores and strength scores. In analysing these graphs, we see that:**
- **It is much more common for a hero to received a high intelligence score than it is**
- **Very few heroes score poorly in intelligence, with most in the 60-100 range.**
- **There are many heroes who, surprisingly, do not score well in strength**

```python
mcu2= mcu1.copy().drop(columns=['has_mind_control_resistance', 'has_mind_control', 'has_durability', 'has_stamina', 'has_reflexes', 'power_score', 'has_endurance', 'has_peak_human_condition'])
```

```python
corr = mcu2.corr()
heat= sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True, cmap=sns.diverging_palette(220, 20, as_cmap=True))
heat.set_title("Relationship of Various Abilities & Score")
```

```python
corr1= mcu1.plot(kind='scatter', x="overall_score", y="strength_score", color='red', figsize=(20,10))
corr1.set_title("Relationship Between Strength Score & Overall Score")
```

- **If you have a low strength score, you are likely to score low overall**

```python
corr2= mcu1.plot(kind='scatter', x="overall_score", y="intelligence_score", figsize=(20,10))
corr2.set_title("Relationship Between Strength Score & Intelligence Score")
```

```python
corr3= mcu1.plot(kind='scatter', x="overall_score", y="intelligence_strength_sum", color='green', figsize=(20,10))
corr3.set_title("Relationship Between Strength Score & Intelligence/Strength Sum")
```

- **Those who have a low overall score still likely have a high intelligence score**

**This data has shown that the age old question of Brains vs Brawns is not a matter of weighting, as I had once thought; both physical aspects and intellectual aspects affect the overall score of a superhero in similar ways. However, through these visualizations, we can see that it is much more common for a superhero to possess intelligence than strength, which is opposite of what my initial thought had been. Based off of this data, brawns is a deciding factor of what makes a good superhero not because it has a higher importance, but because it is more rare to excel in.
